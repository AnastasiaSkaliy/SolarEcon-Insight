Closing Price


The closing price is the last price at which the stock is traded during the regular trading day. A stockâ€™s closing price is the standard benchmark used by investors to track its performance over time.

```python
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator, DateFormatter
# Use dark background style
plt.style.use('dark_background')
# List of unique companies in the 'company_name' column
company_list = df['company_name'].unique()

# Set up the plot
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

# Iterate over each company
for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    
    # Filter data for the current company
    company_data = df[df['company_name'] == company]
    
    # Plot the 'Adj Close' column with the original 'Date' values
    plt.plot(company_data['Date'], company_data['Adj Close'])
    
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f"Closing Price of {company}")

    # Set x-axis tick frequency to display years only
    plt.gca().xaxis.set_major_locator(YearLocator())
    plt.gca().xaxis.set_major_formatter(DateFormatter('%Y'))

plt.tight_layout()
plt.show()
```

Volume of Sales


Volume is the amount of an asset or security that changes hands over some period of time, often over the course of a day. For instance, the stock trading volume would refer to the number of shares of security traded between its daily open and close. Trading volume, and changes to volume over the course of time, are important inputs for technical traders.

```python
import matplotlib.dates as mdates

# Set up the plot
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

# Iterate over each company
for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    
    # Filter data for the current company
    company_data = df[df['company_name'] == company]
    
    # Plot the total volume of stock being traded each day
    plt.plot(company_data['Date'], company_data['Volume'])
    
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f"Sales Volume for {company}")

    # Use DateFormatter for proper date formatting
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.tight_layout()
plt.show()
```

The plots of sales volume for shares provide valuable insights into market activity and investor sentiment. What can we see:

Market Activity: Sales volume plots show the level of trading activity for a particular stock over time. High trading volume often indicates increased interest and activity in the stock, while low volume may suggest indifference or lack of interest.

Liquidity: Volume plots provide an indication of a stock's liquidity, or how easily it can be bought or sold without significantly impacting its price. Higher liquidity is generally preferred by investors as it allows for smoother and more efficient trading.

Price Movements Confirmation: Changes in sales volume often precede price movements. For example, a significant increase in volume accompanied by a price increase may suggest bullish sentiment, while a decrease in volume with a price decrease could signal bearish sentiment.

Market Sentiment: Volume plots can also reflect market sentiment and investor confidence. Sustained high volume during a price rally indicates strong bullish sentiment, while high volume during a price decline may indicate panic selling or capitulation.

Confirmation of Trends: Volume plots can confirm or contradict trends observed in price charts. For example, if a stock price is rising but volume is declining, it may signal weakening momentum and potential reversal.

The plot reveals an interesting observation: in three companies, we observe a surge in trading volume during the period when prices fell between 2023 and 2024.

The observation of high volume accompanying a decline in stock prices is significant and can suggest several potential factors at play:

Panic Selling: High volume during a decline in stock prices may indicate panic selling among investors. When investors become fearful of further losses or negative news, they may rush to sell their shares, resulting in increased trading volume.

Market Sentiment Shift: A notable increase in volume during a downturn could signify a shift in market sentiment. It suggests that a large number of investors are actively participating in the market, either by selling off their positions or taking short positions, reflecting a widespread pessimistic outlook on the stock or market.

Technical Signals: High volume during a decline can confirm technical signals such as breakdowns from key support levels or bearish chart patterns. It adds weight to the bearish signal and indicates strong momentum in the direction of the decline.

Corporate Events or News: Significant corporate events or negative news releases can trigger both a decline in stock prices and a surge in trading volume. This could include earnings misses, regulatory issues, product recalls, or management changes.

Profit-Taking: In some cases, high volume during a decline may also be attributed to profit-taking by investors who had previously enjoyed gains in the stock. They may choose to sell their positions to lock in profits before the decline worsens.

2. What was the moving average of the various stocks?


The moving average (MA) is a widely used technical analysis tool that smooths out price data by creating a constantly updated average price. It provides traders and analysts with a clearer picture of the underlying trend by eliminating short-term fluctuations or noise in the data.

In the provided code, the moving averages for different time periods (10, 20, and 50 days) are calculated for each company's stock prices. The moving averages are plotted alongside the actual adjusted closing prices for visual comparison. This allows analysts to observe how the current price compares to historical averages over different time frames.

By plotting the moving averages, we can identify trends and potential reversal points in the stock prices. When the actual prices cross above or below the moving averages, it may indicate a change in trend direction. Additionally, the relative positions of different moving averages can provide insights into the strength and duration of the prevailing trend.

```python
# Set 'Date' column as the index
df.set_index('Date', inplace=True)

# List of unique companies in the 'company_name' column
company_list = df['company_name'].unique()

# List of moving average days
ma_days = [10, 20, 50]

# Calculate moving averages for each company
for ma in ma_days:
    for company in company_list:
        column_name = f"MA for {ma} days"
        df.loc[df['company_name'] == company, column_name] = df.loc[df['company_name'] == company, 'Adj Close'].rolling(ma).mean()

# Plotting
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

for i, company in enumerate(company_list, 1):
    ax = axes[(i - 1) // 2, (i - 1) % 2]
    columns_to_plot = ['Adj Close'] + [f"MA for {ma} days" for ma in ma_days]
    df[df['company_name'] == company][columns_to_plot].plot(ax=ax)
    ax.set_title(company)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

fig.tight_layout()
plt.show()
```

We see in the graph that the best values to measure the moving average are 10 and 20 days because we still capture trends in the data without noise.
This insight suggests that these shorter-term moving averages may be more suitable for identifying meaningful patterns and trends in the stock prices.

3. What was the daily return of the stock on average?

Now that we've done some baseline analysis, let's go ahead and dive a little deeper. We're now going to analyze the risk of the stock. In order to do so we'll need to take a closer look at the daily changes of the stock, and not just its absolute value.
The daily return of a stock on average refers to the average percentage change in the stock's price from one day to the next over a specific period of time. It is calculated by taking the average of the daily percentage changes in the stock's price over that period.

```python
# Calculate daily returns for each company
for company in company_list:
    df.loc[df['company_name'] == company, 'Daily Return'] = df.loc[df['company_name'] == company, 'Adj Close'].pct_change()

# Plotting
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

for i, company in enumerate(company_list, 1):
    ax = axes[(i - 1) // 2, (i - 1) % 2]
    df[df['company_name'] == company]['Daily Return'].plot(ax=ax, legend=True, linestyle='--', marker='o')
    ax.set_title(company)

fig.tight_layout()
plt.show()
```

We can learn from this plot:
Volatility: The extent of fluctuation or variability in the daily returns indicates the volatility of the stock. A wider range of values indicates higher volatility, while a narrower range suggests lower volatility.

Trends: Patterns in the daily returns can reveal trends in the stock's performance over time. Positive daily returns indicate days when the stock price increased, while negative returns indicate days when the stock price decreased. 

Outliers: Extreme values in the daily returns may represent outliers, indicating days with significant price movements or unusual market events. These outliers can provide insights into specific events or news affecting the stock's performance.

Correlation: Examining the daily returns of multiple companies simultaneously can reveal correlations or relationships between their performances. Positive correlations suggest that the stocks tend to move in the same direction, while negative correlations indicate opposite movements.

For instance, in Enphase Energy, we observe outliers during periods of surges in stock prices. The plots for SolarEdge and SunPower exhibit similarities, depicting numerous days with significant price movements. It indicates for high correlation between these companies. In the case of First Solar, the largest unusual market events are noticeable in 2018 and the first part of 2023.

5. How much value do we put at risk by investing in a particular stock?

There are many ways we can quantify risk, one of the most basic ways using the information we've gathered on daily percentage returns is by comparing the expected return with the standard deviation of the daily returns.

```python
import numpy as np

rets = pivot_df.dropna()

area = np.pi * 20

plt.figure(figsize=(10, 8))
plt.scatter(rets.mean(), rets.std(), s=area)
plt.xlabel('Expected return')
plt.ylabel('Risk')

for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(label, xy=(x, y), xytext=(50, 50), textcoords='offset points', ha='right', va='bottom', 
                 arrowprops=dict(arrowstyle='-', color='blue', connectionstyle='arc3,rad=-0.3'))
```

On this plot, each point represents a different stock. The x-axis indicates the expected return, which is the mean of the daily percentage returns for each stock. The y-axis represents the risk, which is measured by the standard deviation of the daily percentage returns.

By analyzing this plot, we can observe the risk-return tradeoff for each stock.
We can observe that Enphase Energy exhibits the highest expected returns, accompanied by the highest level of risk. Conversely, stocks of First Solar demonstrate lower expected returns but also lower risk. SunPower's stocks are characterized by a low expected return coupled with very high risk. Finally, SolarEdge exhibits a moderate level of both expected return and risk.
