Closing Price


The closing price is the last price at which the stock is traded during the regular trading day. A stocks closing price is the standard benchmark used by investors to track its performance over time.

```python
from matplotlib.dates import YearLocator, DateFormatter
# Use dark background style
plt.style.use('dark_background')
# List of unique companies in the 'company_name' column
company_list = df['company_name'].unique()

# Set up the plot
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

# Iterate over each company
for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    
    # Filter data for the current company
    company_data = df[df['company_name'] == company]
    
    # Plot the 'Adj Close' column with the original 'Date' values
    plt.plot(company_data['Date'], company_data['Adj Close'])
    
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f"Closing Price of {company}")

    # Set x-axis tick frequency to display years only
    plt.gca().xaxis.set_major_locator(YearLocator())
    plt.gca().xaxis.set_major_formatter(DateFormatter('%Y'))

plt.tight_layout()
plt.show()
```

In all the companies, we can observe a significant rise in 2020 and a few years of peak of the green energy industry. However, in the last year (2023) until today (February 2024), there is a colossal decline in the industry, as evident from the shares of key companies. Volume of Sales


Volume is the amount of an asset or security that changes hands over some period of time, often over the course of a day. For instance, the stock trading volume would refer to the number of shares of security traded between its daily open and close. Trading volume, and changes to volume over the course of time, are important inputs for technical traders.

```python
import matplotlib.dates as mdates

# Set up the plot
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

# Iterate over each company
for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    
    # Filter data for the current company
    company_data = df[df['company_name'] == company]
    
    # Plot the total volume of stock being traded each day
    plt.plot(company_data['Date'], company_data['Volume'])
    
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f"Sales Volume for {company}")

    # Use DateFormatter for proper date formatting
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.tight_layout()
plt.show()
```

2. What was the moving average of the various stocks?


The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 10 days, 20 minutes, 30 weeks, or any time period the trader chooses.

```python
# Set 'Date' column as the index
df.set_index('Date', inplace=True)

# List of unique companies in the 'company_name' column
company_list = df['company_name'].unique()

# List of moving average days
ma_days = [10, 20, 50]

# Calculate moving averages for each company
for ma in ma_days:
    for company in company_list:
        column_name = f"MA for {ma} days"
        df.loc[df['company_name'] == company, column_name] = df.loc[df['company_name'] == company, 'Adj Close'].rolling(ma).mean()

# Plotting
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

for i, company in enumerate(company_list, 1):
    ax = axes[(i - 1) // 2, (i - 1) % 2]
    columns_to_plot = ['Adj Close'] + [f"MA for {ma} days" for ma in ma_days]
    df[df['company_name'] == company][columns_to_plot].plot(ax=ax)
    ax.set_title(company)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

fig.tight_layout()
plt.show()
```

We see in the graph that the best values to measure the moving average are 10 and 20 days because we still capture trends in the data without noise.

3. What was the daily return of the stock on average?

Now that we've done some baseline analysis, let's go ahead and dive a little deeper. We're now going to analyze the risk of the stock. In order to do so we'll need to take a closer look at the daily changes of the stock, and not just its absolute value. Let's go ahead and use pandas to retrieve teh daily returns for the Apple stock.

```python
# Calculate daily returns for each company
for company in company_list:
    df.loc[df['company_name'] == company, 'Daily Return'] = df.loc[df['company_name'] == company, 'Adj Close'].pct_change()

# Plotting
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

for i, company in enumerate(company_list, 1):
    ax = axes[(i - 1) // 2, (i - 1) % 2]
    df[df['company_name'] == company]['Daily Return'].plot(ax=ax, legend=True, linestyle='--', marker='o')
    ax.set_title(company)

fig.tight_layout()
plt.show()
```

Great, now let's get an overall look at the average daily return using a histogram.

4. What was the correlation between different stocks closing prices?


We want to analyze the returns of all the stocks in our list. We can compare the daily percentage return of two stocks to check how correlated.

```python
# Drop NaN values in 'Daily Return' column
df_cleaned = df.dropna(subset=['Daily Return'])

# Group by 'Date' and 'company_name', calculate mean daily return
grouped_df = df_cleaned.groupby(['Date', 'company_name'])['Daily Return'].mean().reset_index()
grouped_df.head()
```

```python
# Pivot the DataFrame to create the new structure for dailyreturn
pivot_df = grouped_df.pivot_table(index='Date', columns='company_name', values='Daily Return')

# Display the new DataFrame
print(pivot_df.head())
```

```python
grouped_close_df = df_cleaned.groupby(['Date', 'company_name'])['Adj Close'].mean().reset_index()

# Pivot the DataFrame to create the new structure for dailyreturn
closing_df = grouped_close_df.pivot_table(index='Date', columns='company_name', values='Adj Close')

# Display the new DataFrame
print(closing_df.head())
```

5. How much value do we put at risk by investing in a particular stock?

There are many ways we can quantify risk, one of the most basic ways using the information we've gathered on daily percentage returns is by comparing the expected return with the standard deviation of the daily returns.

```python
import numpy as np

rets = pivot_df.dropna()

area = np.pi * 20

plt.figure(figsize=(10, 8))
plt.scatter(rets.mean(), rets.std(), s=area)
plt.xlabel('Expected return')
plt.ylabel('Risk')

for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(label, xy=(x, y), xytext=(50, 50), textcoords='offset points', ha='right', va='bottom', 
                 arrowprops=dict(arrowstyle='-', color='blue', connectionstyle='arc3,rad=-0.3'))
```
